One of the main ways to filter out rows in a Pandas data frame object is through boolean masking. With boolean masking, certain fucntions (like .loc(), .isin(), df['col'] == , etc.) assign true/false values to each row of the data frame. As a result, when the df[] command is called with a boolean mask, all the rows set to false will be filtered out based on a condition.

## Problem 2:

Plot a simple straight line that goes through (0,3) and (6, 1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


x = np.array([0,6])
y = np.array([3,1])

df = pd.DataFrame({'x': x, 'y' : y})

plt.plot(df['x'], df['y'], )
plt.title('Group 5\'s Awesome Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```
